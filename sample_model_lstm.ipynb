```python
import pretty_midi as pm
import numpy as np
import random
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
```

```python
def midi2NP(midi_file):
    import pretty_midi as pm
    import numpy as np
    music = pm.PrettyMIDI(midi_file)
    output = []
    for track in music.instruments:
        track_output = []
        for note in track.notes:
            track_output.append([note.start, note.end, note.pitch])
        track_output = np.array(track_output)
        output.append(track_output)
    
    return output
```

```python
moonlight = midi2NP('Piano Sonata n14 op27 1mov \'\'Moonlight\'\'.mid')
```

```python
def make_training_data(list_of_NPs, seq_length):
    """
    takes a list of NPs
    returns X, y with correct shape
    """
    seq = seq_length + 1 # account for 1 predicted note
    NP = list_of_NPs[random.randint(0,len(list_of_NPs))]
    seq_start = random.randint(0, NP.shape[0]-seq)
    # actually seq_start + seq - 1, but use python end
    seq_end = seq_start + seq
    sampled_notes = NP[seq_start:seq_end,:]
    X = sampled_notes[:-1,:]
    y = sampled_notes[1:,:]
    return X, y
```

```python
X, y = make_training_data(moonlight, 400)
```

```python
lstm_model = Sequential()
lstm_model.add(LSTM(64, input_shape=(X.shape[1], 1), return_sequences=True))
lstm_model.add(Dropout(0.2))
lstm_model.add(LSTM(64))
lstm_model.add(Dropout(0.2))
lstm_model.add(Dense(y.shape[1], activation='linear'))
lstm_model.compile(loss='mse', optimizer='adam')
```

```python
# define the checkpoint
filepath="tmp/lstm/weights-improvement-{epoch:02d}-{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
```

```python
lstm_model.fit(X.reshape((400,3,1)), y, epochs=10000, batch_size=400, callbacks=callbacks_list)
```

```python
def generate_notes(model, seed):
    output = []
    cur_note = seed
    # generate characters
    for i in range(200):
        x = np.reshape(cur_note, (1, seed.shape[0], 1))
        prediction = model.predict(x, verbose=0)
        prediction[:,2] = np.round(prediction[:,2])
        output.append(prediction)
        cur_note = prediction
    return output
```

```python
generated_song = generate_notes(lstm_model, X[0,:])
```

```python
# reassemble midi file

def notes2Midi(notes, output_file_path):
    start_time = notes[0][:,0].item()
    # create PrettyMIDI object
    output = pm.PrettyMIDI()
    # Use piano as the instrument
    instr = pm.Instrument(0, name = 'Piano')
    # loop over notes
    for note in notes:
        flattened_note = note.flatten()
        note_to_add = pm.Note(velocity=100, pitch=int(flattened_note[2]), start=flattened_note[0]-start_time, end=flattened_note[1]-start_time)
        instr.notes.append(note_to_add)
    output.instruments.append(instr)
    output.write(output_file_path)
```

```python
notes2Midi(generated_song, 'sample_lstm.mid')
```
